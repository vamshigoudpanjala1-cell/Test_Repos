**Palindrome Checker**

**Task: Write a function is_palindrome(text) that returns True if the string is the same when reversed, otherwise False.**

````python
# Solution:
def is_palindrome(text):
    # Convert to lowercase to make the check case-insensitive
    text = text.lower()
    return text == text[::-1]

# Example tests
print(is_palindrome("madam"))      # True
print(is_palindrome("Racecar"))    # True
print(is_palindrome("hello"))      # False
````

**Multiple Return Values**

**Task: Create a function circle_measurements(radius) that returns both area and circumference of a circle.**

**Use formulas:**

**Area = π × r²**

**Circumference = 2 × π × r**

````python
# Solution:
import math

def circle_measurements(radius):
    area = math.pi * radius ** 2
    circumference = 2 * math.pi * radius
    return area, circumference

# Example usage
area, circumference = circle_measurements(5)

print(f"Area: {area}")
print(f"Circumference: {circumference}")
````

**Task: Create a function that finds both the smallest and largest of three numbers.**
```
Output:

Smallest: 5, Largest: 12
Range: 25 to 100
```

````python
# Solution:
def min_max(num1, num2, num3):
    smallest = min(num1, num2, num3)
    largest = max(num1, num2, num3)
    return smallest, largest

# Example usage
smallest, largest = min_max(12, 7, 25)

print(f"Smallest number: {smallest}")
print(f"Largest number: {largest}")
````

**Task: Create a function that takes a full name and returns first name and last name separately.**

**Output:**
```
Name Amrita Prasad
First name: Amrita, Last name: Prasad
Hello Amrita! Your last name is Prasad.

````python
# Solution:
def split_name(full_name):
    parts = full_name.strip().split()
    if len(parts) >= 2:
        first_name = parts[0]
        last_name = parts[-1]
    else:
        # If only one name is provided, last name is empty
        first_name = parts[0]
        last_name = ""
    return first_name, last_name

# Example usage
first, last = split_name("vamshi goud")
print("First Name:", first)
print("Last Name:", last)
````

**Task: Create a function that finds the highest number from any amount of scores**

````python
# Solution:
def highest_score(*scores):
    if not scores:
        return None  # Return None if no scores are provided
    return max(scores)

# Example usage
print("Highest score:", highest_score(85, 92, 78, 99, 67))
print("Highest score:", highest_score(50, 40))
print("Highest score:", highest_score())  # No scores provided
````

**Task: Create a function that calculates the average of any amount of numbers**

````python
# Solution:
def average(*numbers):
    if not numbers:
        return None  # Return None if no numbers are provided
    return sum(numbers) / len(numbers)

# Example usage
print("Average:", average(10, 20, 30))         # 20.0
print("Average:", average(5, 15, 25, 35, 45))  # 25.0
print("Average:", average())                    # None
````

**Function with Keyword Arguments :**

**Task: Create a function book_info(title, author, year=2025) that prints the book details. Call it using positional and keyword arguments.**

````python
# Solution:
def book_info(title, author, year=2025):
    print(f"Title: {title}")
    print(f"Author: {author}")
    print(f"Year: {year}")

# Calling with positional arguments
book_info("The Great Adventure", "Amrita Prasad", 2023)

print("----------------------")

# Calling with keyword arguments
book_info(title="Python Fundamentals", author="John Doe", year=2024)

print("----------------------")

# Calling with default year
book_info(title="Data Analytics 101", author="Alice Smith")
````
